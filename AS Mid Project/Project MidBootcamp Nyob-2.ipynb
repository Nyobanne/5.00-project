```python
# using Atwater factors (per kilogram)
atwater_factors = {
    'Protein': 17,      # 17 kJ/g 
    'Carbohydrate': 17, # 17 kJ/g 
    'Fat': 37,          # 37 kJ/g 
    'Sugars': 17,       # 17 kJ/g 
}

# convert g/kg to kJ/kg
total_energy_kj = (
    df['Protein (g/kg)'] * atwater_factors['Protein'] +
    df['Carbohydrate (g/kg)'] * atwater_factors['Carbohydrate'] +
    df['Fat (g/kg)'] * atwater_factors['Fat'] +
    df['Sugars (g/kg)'] * atwater_factors['Sugars']
)

# Assign calculated values to missing values in 'Energy, Regulation EU No 1169/2011 (kJ/kg)'
df['Energy, Regulation EU No 1169/2011 (kJ/kg)'].fillna(total_energy_kj, inplace=True)
df.rename(columns={'Energy, Regulation EU No 1169/2011 (kJ/kg)': 'Energy (kJ/kg)'}, inplace=True)
```

```python
from scipy.stats import norm, expon
# Set up the subplots
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(16, 8))
fig.suptitle('Distribution Plots before Transformation', y=1.02)

# Data is continuous so I can make violinplots for each column
for col, ax in zip(columns_to_plot, axes[1]):
    sns.violinplot(data=df, x=col, ax=ax, orient='v', color='lightcoral', width=0.3)
    ax.set_title(f'Violin plot of {col}')
    ax.set_xlabel('')
    ax.set_ylabel('')
    
# Create distribution plots for each column
for col, ax in zip(columns_to_plot, axes.flatten()):
    sns.histplot(df[col], kde=True, ax=ax, stat='density')  # Use stat='density' for probability density estimation
    
    # Add theoretical normal curve
    x = np.linspace(df[col].min(), df[col].max(), 1000)
    normal_curve = norm.pdf(x, loc=df[col].mean(), scale=df[col].std())
    ax.plot(x, normal_curve, color='red', label='Theoretical Normal Curve')
    
    # Add theoretical exponential curve
    exponential_curve = expon.pdf(x, scale=df[col].mean())  # Exponential distribution with mean as scale parameter
    ax.plot(x, exponential_curve, color='green', label='Theoretical Exponential Curve')
    
    ax.set_title(f'Distribution of ({col})')
    ax.legend()

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Set display options to show all rows and columns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

correlation_matrix = numerical.corr(method='pearson')
filtered_correlation = correlation_matrix[((correlation_matrix > 0.2) | (correlation_matrix < -0.2))&(correlation_matrix < 1) ]

# Stack the filtered values into a DataFrame
filtered_correlation_stacked = filtered_correlation.stack()

# Create a DataFrame from the stacked values
summary_table = pd.DataFrame(filtered_correlation_stacked, columns=['Correlation'])

# Print the summary table
print("Summary Table of Correlations Above 0.2 or Below -0.2:")
display(summary_table)
# Reset display options to default
pd.reset_option('display.max_rows')
pd.reset_option('display.max_columns')
```

### Comments: 
#### ENVIRONMENT (env)
- EF Score is dependant of all the other (env) environmental indicators.
- EF_Score is very correlated to Carbon, therefore the (nut)/(env) associations are similar
- Farming stage and Total are very correlated
- What happens through stage? let's have a closer look at EF_Score:
    - EF_Score Farm +(Protein, Zinc,B3)
    - EF_Score Process+(Mg,K,B3)
    - EF_Score Pack: not related to (nut)
    - EF_Score Distrib	+(Protein, -Carb, -Sugars, Zinc) : and also a larger number of (env).
    - EF_Score Consum +(Protein, Zinc, B3, -Sugars)
    - EF_Score Total +(Protein, Zinc, B3)

#### NUTRITIONAL (nut)
- Vitamins B are all dependant of each other. B3 would be a good one to keep. 

#### DQR
- The Higher the DQR, the worse is data quality.
- DQR is highest with; EF transp, NMVOC, Particulates, Sweet waterAcid and Eutroph, Water Use. therefore those will be deprioritized for this analysis

#### Food groups
- Food Group and Subgroup have associations with Sweet Water impact, Earth Eutroph


#### intermediate conclusion
- surprisingly,(nut) Energy, Fat and Saturated Fat are hardly associated with (env)
- Protein, B3 and Zinc show the most frequent positive association
- Carbs and Sugars show the most frequent negative association
- Later on we should check if the association is due to animal/vegetable nature of source foods

# a) model nut/env


```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

# Define features (X) and target variable (y)
X = data[['Carbohydrate (g/kg)', 
          'Protein (g/kg)', 'Magnesium (mg/kg)', 
          'Potassium (mg/kg)']]
y = data['EF_Score Total']

# Scale the features using Min-Max scaling
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=40)

# Initialize the Linear Regression model
model = LinearRegression()

# Train the model on the training data
model.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = model.predict(X_test)
```

#### Model Validation

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print('Mean Squared Error (MSE):', mse) #accuracy
print('R-squared (R2):', r2) #proportion of the variance in the dependent variable is explained by the model

# Optional: Model Interpretation
coefficients = model.coef_
print('Coefficients:', coefficients)
```
